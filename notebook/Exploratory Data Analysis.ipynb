# Explorattory Data Analysis (EDA)

```python
#Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the clean dataset
df = pd.read_csv("../data/clean/clean_cobify.csv")
df.head()
```

```python
df_num = df.select_dtypes(include=['number'])
df_cat = df.select_dtypes(include=['object'])
```

```python
#Plot a graph displaying the frequency of each gas_type
gas_type_counts = df['gas_type'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=gas_type_counts.index, y=gas_type_counts.values, palette="viridis")
plt.title('Count of Each Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

```python
for col in df_num:
    sns.displot(df_num[col])
    plt.show()
```

```python
e10 = df[df['gas_type'] == 'E10']
sp98 = df[df['gas_type'] == 'SP98']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharex=True, sharey=True)

ax1.hist(e10['distance'], bins=50)
ax1.set_xlabel('Distance')
ax1.set_ylabel('Number of Trips')
ax1.set_title('E10')

ax2.hist(sp98['distance'], bins=50)
ax2.set_xlabel('Distance')
ax2.set_ylabel('Number of Trips')
ax2.set_title('SP98')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharex=True, sharey=True)

ax1.hist(e10['cost_of_trip'], bins=50)
ax1.set_xlabel('Cost of trips')
ax1.set_ylabel('Number of Trips')
ax1.set_title('E10')

ax2.hist(sp98['cost_of_trip'], bins=50)
ax2.set_xlabel('Cost of trips')
ax2.set_ylabel('Number of Trips')
ax2.set_title('SP98')

plt.show()
```

```python
plt.figure(figsize=(20, 18))

# Plot for E10
sns.lineplot(data=e10, x='speed', y='cost_of_trip', color='blue', label='E10')

# Plot for SP98
sns.lineplot(data=sp98, x='speed', y='cost_of_trip', color='green', label='SP98')

# Set labels and title
plt.xlabel('Average speed')
plt.ylabel('Cost of trip')

# Display the plot
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 18))

# Plot for E10
sns.lineplot(data=e10, x='distance', y='fuel_consumed', color='blue', label='E10')

# Plot for SP98
sns.lineplot(data=sp98, x='distance', y='fuel_consumed', color='green', label='SP98')

# Set labels and title
plt.xlabel('Distance Travelled')
plt.ylabel('Fuel Consumed')

# Display the plot
plt.legend()
plt.show()
```

### Insights

- We see that when travelling at a higher average speed, then the E10 fuel type is more efficient and in its fuel consumption and reduces the cost of the trip. Higher average speeds can indicate longer trips.
- There does not seem to be a significant difference in cost of trips or fuel consumption for shorter distance trips
- I would recommend using E10 fuel type due to its cheaper price and higher efficiency in its consumption. This particularly holds true for longer distances or trips with a higher average speed.
